# **Image Shearing**

Write a Python program using OpenCV to perform image shearing on the **given image**.

The program should:

1. Create a **shearing transformation matrix** to apply:
   - **Horizontal shearing**  
   - **Vertical shearing**
2. Apply the shearing transformations.
3. Display the following using a **1 × 3 layout** in Matplotlib:  
   - **Original Image**  
   - **Horizontally Sheared Image**  
   - **Vertically Sheared Image**


```python
# Horizontal Shearing

horizontal_matrix = np.float32([[1, 0.5, 0],[0, 1, 0]])
horizontal_shear = cv2.warpAffine(image,horizontal_matrix,(636, 438))
```

```python
# vertical shearing
vertical_matrix = np.float32([[1, 0, 0],[0.5, 1, 0]])
vertical_shear = cv2.warpAffine(image,vertical_matrix,(636, 438))

```

```python
plt.figure(figsize=(15,5))

# Original Image
plt.subplot(1,3,1)
plt.imshow(image[:, :, ::-1])
plt.title("Original Image")
plt.axis('off')

# Horizontally Sheared Image
plt.subplot(1,3,2)
plt.imshow(horizontal_shear[:, :, ::-1])
plt.title("Horizontally Sheared Image")
plt.axis('off')

# Vertically Sheared Image
plt.subplot(1,3,3)
plt.imshow(vertical_shear[:, :, ::-1])
plt.title("Vertically Sheared Image")
plt.axis('off')

plt.tight_layout()
plt.show()
```

# **Image Rotation**

Write a Python program using OpenCV to perform **image rotation** on the **given image**.

The program should:

1. Rotate the image by:  
   - **45 degrees**  
   - **90 degrees**
2. Use `cv2.getRotationMatrix2D` and `cv2.warpAffine`.
3. Display the following using a **1 × 3 layout** in Matplotlib:  
   - **Original Image**  
   - **45° Rotated Image**  
   - **90° Rotated Image**


```python
# 45 degrees rotation
(height, width) = image.shape[:2]  
angle = 45  
center = (width // 2, height // 2) 
M_rotation = cv2.getRotationMatrix2D(center, angle, 1)  
rotated_image = cv2.warpAffine(image, M_rotation, (width, height))  
```

```python
# 90 degree rotations
(height, width) = image.shape[:2]  
angle = 90  
center = (width // 2, height // 2) 
M_rotation = cv2.getRotationMatrix2D(center, angle, 1)  
rotated_image2 = cv2.warpAffine(image, M_rotation, (width, height))  
```

```python
plt.figure(figsize=(15,5))

# Original Image
plt.subplot(1,3,1)
plt.imshow(image[:, :, ::-1])
plt.title("Original Image")
plt.axis('off')

# Horizontally Sheared Image
plt.subplot(1,3,2)
plt.imshow(rotated_image[:, :, ::-1])  # Display the rotated image
plt.title(" 45 degrees Rotated Image")  # Set title
plt.axis('off')

# Vertically Sheared Image
plt.subplot(1,3,3)
plt.imshow(rotated_image2[:, :, ::-1])  # Display the rotated image
plt.title(" 90 degrees Rotated Image")  # Set title
plt.axis('off')

plt.tight_layout()
plt.show()
```
